## Creación de la red

#### Actividad 1 Complete la siguiente lista de enlaces para representar la red de amistades de la Practica 1. Python distingue entre mayúsculas y minúsculas.

```python
lista_enlaces = [
    ("Alicia","Beto"),
    ("Carlos", "Alicia"),
    ("Alicia","David"),
    ("Alicia", "Carmen"),
    ("Alicia","Monica"),
    ("Beto", "Jorge"),
    ("Jorge","Enrique"),
    ("Enrique", "Rosa"),
    ("Rosa","Jorge"),
    ("Enrique", "Irene"),
    ("Irene","Jorge"),
    ("Irene", "Rosa"),
    ("Carmen","Monica"),
    ("David", "Carlos"),
    ("Carlos","Monica"),
    ("Carmen", "Irene")
]
```

## Función para mostrar los caminos (más cortos) y distancia

La siguiente funcion recibe como parametros los nodos: nodo1 y nodo2 e imprime los caminos mas cortos entre ellos. La funcion devuelve la distancia entre el nodo1 y el nodo2

```python
def cam_dist(nodo1,nodo2):
    """
    Recibe como parametros los nodos: nodo1 y nodo2
    Imprime los caminos mas cortos entre nodo1 y nodo2 
    Devuelve la distancia entre el nodo1 y el nodo2
    """
    caminos = nx.all_shortest_paths(G, nodo1, nodo2)
    for camino in caminos: 
        print("Uno de los caminos más corto entre " + nodo1 + " y " + nodo2 +" es " + str(camino))
    distancia = nx.shortest_path_length(G, nodo1, nodo2)
    return distancia
```

Utilizamos las funciones para mostrar caminos y distancias

```python
dist_BetoCarlos = cam_dist("Beto","Carlos")
dist_BetoCarlos
```

## Función para calcular la distancia promedio y diametro de una red

Actividad 7 Cree una función recibe como parametros de una red G, que imprima el diámetro y la distancia promedio de G. La función debe devolver la distncia promedio de G.

```python
def dandd(G):
    diametro=nx.diameter(G)
    print("El diametro de la red es: "+str(diametro))
    prom_dist=nx.average_shortest_path_length(G)
    print("La distancia promedio es: "+str(prom_dist))
    return prom_dist
```
